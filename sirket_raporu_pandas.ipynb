```python
import pandas as pd
df = pd.read_csv("sirket.csv")
ozet=df[["maas","yas","performans","calisma_yili"]].agg(["sum","mean"]).round(2)
en_yuksek_maas=df.loc[df["maas"].idxmax()]
en_dusuk_maas=df.loc[df["maas"].idxmin()]
en_yuksek_performans=df.loc[df["performans"].idxmax()]
en_dusuk_performans=df.loc[df["performans"].idxmin()]
departman_sayim=df.groupby("departman").size()
istanbul= df[df["sehir"]=="İstanbul"]

with open("rapor_pandas.txt", "w", encoding="utf-8") as f:
    f.write("=== ŞİRKET RAPORU ===\n\n")
    
    f.write("--- Genel İstatistikler ---\n")
    f.write(ozet.to_string())                          # tabloyu string'e çevir
    f.write(f"\n\nToplam Çalışan: {len(df)}\n")
    
    f.write("\n--- Maaş Uçları ---\n")
    f.write(f"En Yüksek: {en_yuksek_maas['isim']} - {en_yuksek_maas['maas']} TL\n")
    f.write(f"En Düşük: {en_dusuk_maas['isim']} - {en_dusuk_maas['maas']} TL\n")
    
    f.write("\n--- Performans Uçları ---\n")
    f.write(f"En Yüksek: {en_yuksek_performans['isim']} - {en_yuksek_performans['performans']}\n")
    f.write(f"En Düşük: {en_dusuk_performans['isim']} - {en_dusuk_performans['performans']}\n")
    
    f.write("\n--- Departman Dağılımı ---\n")
    f.write(departman_sayim.to_string())
    
    f.write("\n\n--- Departman Bazında Ortalama Maaş ---\n")
    f.write(df.groupby("departman")["maas"].mean().round(2).to_string())
    
    f.write("\n\n--- Şehir Bazında Ortalama Maaş ---\n")
    f.write(df.groupby("sehir")["maas"].mean().round(2).to_string())
    
    f.write(f"\n\n--- İstanbul Çalışanları ({len(istanbul)}) ---\n")
    f.write(istanbul[["isim", "yas", "departman"]].to_string())

# Kontrol
with open("rapor_pandas.txt", "r", encoding="utf-8") as f:
    print(f.read())

```
