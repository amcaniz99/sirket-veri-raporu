```python
sirket=[]
with open("sirket.csv" , "r", encoding="utf-8") as f:
    basliklar= f.readline().strip().split(",")

    for satir in f:
        degerler= satir.strip().split(",")
        kayit={}
        for i in range(len(basliklar)):
            deger=degerler[i]
            if basliklar[i] in ["yas","maas","calisma_yili"]:
                deger=int(deger)
            elif basliklar[i] in ["performans"]:
                deger=float(deger)
            kayit[basliklar[i]]=deger
        sirket.append(kayit)
        
toplam_maas=0
toplam_yas=0
toplam_performans=0
toplam_calisma_yili=0
en_yuksek_maas=sirket[0]
en_dusuk_maas=sirket[0]
en_yuksek_performans=sirket[0]
en_dusuk_performans=sirket[0]
departman_sayim={}
istanbul_calisanlari=[]

for s in sirket:
    toplam_yas+=s["yas"]
    toplam_maas+=s["maas"]
    toplam_performans+=s["performans"]
    toplam_calisma_yili+=s["calisma_yili"]
    if s["maas"]> en_yuksek_maas["maas"]:
        en_yuksek_maas=s
    if s["maas"]<en_dusuk_maas["maas"]:
            en_dusuk_maas=s
    if s["performans"]>en_yuksek_performans["performans"]:
        en_yuksek_performans=s
    if s["performans"]<en_dusuk_performans["performans"]:
        en_dusuk_performans=s
    dept=s["departman"]
    if dept in departman_sayim:
        departman_sayim[dept]+=1
    else:
        departman_sayim[dept]=1
    if s["sehir"]=="İstanbul":
        istanbul_calisanlari.append(s)

    

ortalama_maas= toplam_maas/len(sirket)
ortalama_yas=toplam_yas/len(sirket)
ortalama_performans= toplam_performans/len(sirket)
ortalama_calisma_yili= toplam_calisma_yili/len(sirket)

with open ("rapor2.txt" , "w", encoding="utf-8") as f:
    f.write(f" === SIRKET RAPORU ===\n")
    f.write(f"Toplam Calisan : {len(sirket)}\n")
    f.write(f"Toplam Maas: {round(toplam_maas,2)} TL\n")
    f.write(f"Ortalama Maas: {round(ortalama_maas,2)} TL\n")
    f.write(f"Ortalama Yas: {round(ortalama_yas,2)}\n")
    f.write(f"Ortalama Performans: {round(ortalama_performans,2)}\n")
    f.write(f"Ortalama Calisma Yili: {round(ortalama_calisma_yili,2)}\n")

    f.write(f" --- Maas Uclari ---\n")
    f.write(f"En Yuksek Maas: {en_yuksek_maas['isim']} - {en_yuksek_maas['maas']} TL\n")
    f.write(f"En Dusuk Maas: {en_dusuk_maas['isim']} - {en_dusuk_maas['maas']} TL\n")

    f.write(f" --- Performans Uclari ---\n")
    f.write(f"En Yuksek Performans: {en_yuksek_performans['isim']} - {en_yuksek_performans['performans']}\n")
    f.write(f"En Dusuk Performans: {en_dusuk_performans['isim']} - {en_dusuk_performans['performans']}\n")

    f.write(f" --- Departman Dagilimi ---\n")
    for dept, sayi in departman_sayim.items():
        f.write(f"{dept}: {sayi} kişi\n")

    f.write(f"--- Sehir Filtreleri ---\n")
    f.write(f"İstanbul'daki Çalışanlar ({len(istanbul_calisanlari)}):\n")
    for c in istanbul_calisanlari:
        f.write(f"- {c['isim']} ({c['yas']}, {c['departman']})\n")



with open("rapor2.txt", "r", encoding="utf_8") as f:
    print(f.read())
```
