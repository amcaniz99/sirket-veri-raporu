```python
print("--- İstanbul'daki çalışanlar ---")
istanbul = df[df["sehir"] == "İstanbul"]
print(istanbul)

print("\n--- Yazılım departmanındaki çalışanlar ---")
yazilim = df[df["departman"] == "Yazılım"]
print(yazilim)

print("\n--- Maaşı 20000'den yüksek olanlar ---")
yuksek_maaslilar = df[df["maas"] > 20000]
print(yuksek_maaslilar)

print("\n--- Hem İstanbul hem maaşı 20000+ olanlar ---")
kosul = df[(df["sehir"] == "İstanbul") & (df["maas"] > 20000)]
print(kosul)
```

```python
#groupby veriyi bir sütunun değerine göre gruplara böler. Sonra her gruba ayrı hesap yapabilirim — sum, mean, count, min, max gibi. SQL'deki GROUP BY ile aynı. Veri analizinin temeli.
```

```python
df["yillik_maas"]=df["maas"]*12
df["performans_basina_maas"]=df["maas"]/df["performans"]
df["performans_kategori"]=df["performans"].apply(lambda x:"yüksek" if x>=8 else "düşük")
print(df)
```

```python
df=pd.read_csv("sirket.csv")
ozet=df[["maas","yas","performans","calisma_yili"]].agg(["sum","mean"]).round(2)
en_yuksek_maas=df.loc[df["maas"].idxmax()]
en_dusuk_maas=df.loc[df["maas"].idxmin()]
en_yuksek_performans=df.loc[df["performans"].idxmax()]
en_dusuk_performans=df.loc[df["performans"].idxmin()]
departman_sayim=df.groupby("departman").size()
istanbul= df[df["sehir"]=="İstanbul"]

print("=== SİRKET RAPORU ===\n")
print("--- Genel İstatistikler ---\n")
print(ozet)
print(f"\nToplam Çalışan: {len(df)}")

print("\n--- Maaş Uçları ---")
print(f"En Yüksek: {en_yuksek_maas['isim']} - {en_yuksek_maas['maas']} TL")
print(f"En Düşük: {en_dusuk_maas['isim']} - {en_dusuk_maas['maas']} TL")

print("\n--- Performans Uçları ---")
print(f"En Yüksek: {en_yuksek_performans['isim']} - {en_yuksek_performans['performans']}")
print(f"En Düşük: {en_dusuk_performans['isim']} - {en_dusuk_performans['performans']}")

print("\n--- Departman Dağılımı ---")
print(departman_sayim)

print(f"\n--- İstanbul Çalışanları ({len(istanbul)}) ---")
print(istanbul[["isim", "yas", "departman"]])

print("\n--- Departman Bazında Ortalama Maaş ---")
print(df.groupby("departman")["maas"].mean().round(2))

print("\n--- Şehir Bazında Ortalama Maaş ---")
print(df.groupby("sehir")["maas"].mean().round(2))

```
